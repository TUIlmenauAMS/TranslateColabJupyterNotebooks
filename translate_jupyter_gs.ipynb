```python
import os
import json
import sys
#select file on the left pane and copy the path
filename = '/content/gdrive/MyDrive/Klimawandel_Nachgerechnet/climateChangeCalculated6.ipynb' #@param {'type': 'string'}
#languages: ['en', 'de','fr', 'es', 'hi', 'ja']...
targetlanguage= 'de' #@param ['en', 'de','fr', 'es', 'hi', 'ja']
print("filename:", filename)
print("targetlanguage:",targetlanguage)
filenamewoe, file_extension = os.path.splitext(filename)
#filename_translated='/content/gdrive/MyDrive/Klimawandel_Nachgerechnet/translated.ipynb'
filename_translated=filenamewoe + "_"+ targetlanguage + file_extension
print("filename_translated=", filename_translated)
with open(filename, "r") as f:
  notebook = json.load(f)
```

```python
# Create translation model
translate = Translation()

# Download the JSON file.
with open(filename, "r") as f:
   notebook = json.load(f)
new_notebook = notebook.copy()
cells = notebook["cells"].copy()
new_cells = cells.copy()
offset = 1

for index, cell in enumerate(cells):
  print("===index: " + str(index) )
  
  if cell["cell_type"] == "markdown":
    source=(cell["source"])
    for i in range(len(source)):
      source[i]=source[i].replace('\n','<br>') #replace return before translation
    #print("cell[source]: ", source)
    translation = translate(source, target=targetlanguage)
    for i in range(len(translation)):
      translation[i]=translation[i].replace('<br>','\n') #replace return back after translation
    #print("translated:", translation)
    new_cell = { 'cell_type': 'markdown', 'source': translation }
    #new_cells.insert(index + offset, new_cell) #for interleaving the translated
    #offset = offset + 1
    new_cells[index]= new_cell #for replacing by the translated
    
new_notebook["cells"] = new_cells

# delete meta data
for index in range(0,len(new_notebook["cells"])):
  new_notebook["cells"][index]["metadata"] = {}
  
new_notebook["metadata"] = {}

print("write to "+filename_translated)

with open(filename_translated, "w") as f:
  json.dump(new_notebook, f, ensure_ascii=False, indent=2)
print('Done')
```
